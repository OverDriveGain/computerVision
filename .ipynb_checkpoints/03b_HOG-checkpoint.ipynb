## Gradients and directions
First, implement the extraction of the gradient via convolution with the Sobel kernels. Translate the two result matrices into an image containing the gradient direction (one angle per pixel). Display this image for each of the three input images above. Like it fancy? Add transparency inversely proportional to the gradient magnitude (weak gradients are transparent). 

```python
# dieser Code ist Teil eine Musterlösung, die von Adrian Defer zur Verfügung gestellt und von Tim renoviert wurde

    
def getGradientConv(image):
    Sx = np.array([[-1,0,1],[-1,0,1],[-1,0,1]]) 
    Sy = Sx.T
    Gx = ndimage.convolve(color.rgb2gray(image), Sx)
    Gy = ndimage.convolve(color.rgb2gray(image), Sy)
    GxSqr = np.multiply(Gx, Gx)
    GySqr = np.multiply(Gy, Gy)
    GradSqr = np.add(GxSqr, GySqr)
    Grad = np.sqrt(GradSqr)
    mean = np.mean(Grad)
    threshIndices = Grad < 1.7*mean #Threshold
    Grad[threshIndices] = 0
    return Grad
    # ...
    # return ...


fig = plt.figure()
ax1 = plt.subplot(1, 3, 1)
ax2 = plt.subplot(1, 3, 2)
ax3 = plt.subplot(1, 3, 3)
ax1.imshow(getGradientConv(pers1))
ax2.imshow(getGradientConv(pers2))
ax3.imshow(getGradientConv(pers3))

# ax1.imshow(...)
# ax2.imshow(...)
# ax3.imshow(...)
```

## Gradient orientation histograms (GOHs)
Now implement a function that generates GOHs on image cells. Pass a gradient direction image (as described above) to the function. The image cells should be 8 x 8 pixels in size. Please do not forget the voting with "bi-linear interpolation" (important!). Compute the GOH for the input image "star.png" with and without interpolation! Now rotate the image by -5° and compute both variants of the GOH again! What are the Euclidean differences of the histograms (original vs rotated) with and without interpolation?


```python
star = io.imread('images/star.png')
def printMy(array):
    isTrue = 0
    if(isinstance(array, (list, np.ndarray))==False):
        if(array!=0):print(array)
        return
    for i in array:
        if isinstance(i, (list, np.ndarray)):
            for u in i:
                if isinstance(i, (list, np.ndarray)) == False:
                    for w in u:
                        if(w != 0):
                            isTrue = 1
                else:
                    if(u != 0):
                        isTrue = 1
        else:
            if(i != 0):
                isTrue = 1
    if(isTrue ==1): print(array)
        
def getWindowGrads(image, x, y, length):
    Sx = np.array([[-1,0,1],[-1,0,1],[-1,0,1]]) 
    Sy = Sx.T
    Gx = ndimage.convolve(color.rgb2gray(image[x:x+length, y:y+length]), Sx)
    Gy = ndimage.convolve(color.rgb2gray(image[x:x+length, y:y+length]), Sy)
    return 180/math.pi* (np.arctan2(Gx, Gy)% math.pi), np.sqrt(np.add(np.multiply(Gy, Gy), np.multiply(Gx, Gx)))
    
def getGradHistogram(windowDirections, windowMagnitueds):
    bins_ = np.arange(10, 190, 20)
    result = np.zeros((9,1))
    itsTrue = 0
    for (idx, i) in enumerate(windowDirections):
        for (udx, u) in enumerate(i):
            for bin_ in range (8,0, -1):
                if(u - bins_[bin_] >= 0):
                     distance = u - bins_[bin_]
                     mue = distance / 20   
                     result[bin_] = result[bin_] + mue* windowMagnitueds[idx,udx]
                     if(bin_+1 > 8):  
                        result[0] = result[0] + (1-mue)* windowMagnitueds[idx,udx]
                     else:
                        result[bin_+1] = result[bin_+1] + (1-mue)* windowMagnitueds[idx,udx]
                     break;
    return result

def getGradHistogramWithoutInt(windowDirections, windowMagnitueds):
    bins_ = np.arange(10, 190, 20)
    result = np.zeros((9,1))
    itsTrue = 0
    for (idx, i) in enumerate(windowDirections):
        for (udx, u) in enumerate(i):
            for bin_ in range (8,0, -1):
                if(u - bins_[bin_] >= 0):
                     result[bin_] = result[bin_] + windowMagnitueds[idx,udx]
                     break;
    return result

def getBlock(array, i, u):
    first = np.add(array[i][u], array[i][u+1])
    second = np.add(array[i+1][u], array[i+1][u+1])
    norm = np.linalg.norm(np.add(first, second))
    if(norm == 0): norm =1
    return np.add(first, second)/norm

HOGFeature = []
HOGFeatureWithoutInter = []

def iterateStar():
    gradWindows = []
    gradWindowsWithoutInter = []
    for i in range ( 0, len(star), 8):
        gradWindows.append([])
        gradWindowsWithoutInter.append([])
        for u in range(0, len(star[i]), 8):
            x,y = getWindowGrads(star, i,u,8)
            gradWindows[-1].append(getGradHistogram(x, y))
            gradWindowsWithoutInter[-1].append(getGradHistogramWithoutInt(x, y))
        
    windowsBlocks = []
    windowsBlocksWithoutInter = []
    for i in range(0, len(gradWindows) -1):
        windowsBlocks.append([])
        windowsBlocksWithoutInter.append([])
        for u in range (0, len(gradWindows[0])-1):
            windowsBlocks[-1].append(getBlock(gradWindows, i, u)) #blocks share 2 cells
            windowsBlocksWithoutInter[-1].append(getBlock(gradWindowsWithoutInter, i, u))

    HOGFeature = np.array((windowsBlocks[0][0]))
    HOGFeatureWithoutInter = np.array((windowsBlocks[0][0]))
    for i in windowsBlocks:
        for u in i:
            HOGFeature = np.add(HOGFeature, u)

    for i in windowsBlocksWithoutInter:
        for u in i:
            HOGFeatureWithoutInter = np.add(HOGFeatureWithoutInter, u)
    return windowsBlocks , HOGFeature, HOGFeatureWithoutInter

windowsBlocks, x, y = iterateStar()

star = ndimage.rotate(star, -5)
windowsBlocks2, x5, y5 = iterateStar()

print('#Euclidean distance after rotation with interpolation')
print(np.linalg.norm(x5-x))
print('# without interpolation')
print(np.linalg.norm(y5-y))
    
```

## Block and ROI normalization and creation of the HOG descriptor
* Given a fixed size ROI (128 x 64 pixels), create a descriptor as shown in the lecture (including normalization of the blocks and the total ROI). Please reuse the precalculated cell histograms for the blocks! 
* Calculate the descriptor for the three images above and for the unknown images "unknown1.png" and "unknown2.png". 
* Now compare the unknowns with the three person instances and calculate the Euclidean distance of the descriptors. Which picture has a smaller distance to the persons? (**RESULT**)

```python
## 128x64 pixels yields 16x8 cells yields 15x7 blocks
def createDesc(blocksHistograms, x=0, y=0):
    HOGFeature = np.array((blocksHistograms[0][0]))
    for i in range (x, x+15):
        for u in range(y , y+7):
            HOGFeature = np.add(HOGFeature, blocksHistograms[i][u])
    return HOGFeature
#Create descriptor reuse precalculated histograms

#first image above
star = pers1 
w,x,y = iterateStar()
pers1desc= createDesc(w)

star = pers2 
w,x,y = iterateStar()
pers2desc= createDesc(w)

star = pers3 
w,x,y = iterateStar()
pers3desc= createDesc(w)

star = io.imread('images/unknown1.png')
w,x,y = iterateStar()
unknown1desc = createDesc(w)

star = io.imread('images/unknown2.png')
w,x,y = iterateStar()
unknown2desc= createDesc(w)

print(np.linalg.norm(np.array(unknown1desc)-np.array(pers1desc)))
print(np.linalg.norm(np.array(unknown1desc)-np.array(pers2desc)))
print(np.linalg.norm(np.array(unknown1desc)-np.array(pers3desc)))
#91.92941234479417
#91.9994484842324
#81.41290165666524

print(np.linalg.norm(np.array(unknown2desc)-np.array(pers1desc)))
print(np.linalg.norm(np.array(unknown2desc)-np.array(pers2desc)))
print(np.linalg.norm(np.array(unknown2desc)-np.array(pers3desc)))
#98.24219163365002
#98.05979027574193
#90.96608979924508

## Obviously unkown1 is closer.
```
